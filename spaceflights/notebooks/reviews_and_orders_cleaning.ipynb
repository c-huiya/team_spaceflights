```python
import pandas as pd
import numpy as np
import os
```

```python
order = pd.read_csv('olist_orders_dataset.csv')
print(order.head())
print(order.info())
```

```python
order.columns
```

```python
order = order.drop_duplicates()
```

```python
num_dups = order.duplicated().sum()
print(num_dups)
```

```python
print(order.isna().sum())
```

```python
# 2. Parse each timestamp column into real datetimes
ts_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
]
for col in ts_cols:
    order[col] = pd.to_datetime(order[col], errors="coerce")

# 3. Create “clean” string versions with two-digit hours/min/sec
for col in ts_cols:
    order[f"{col}_clean"] = order[col].dt.strftime("%Y-%m-%d %H:%M:%S")
```

```python
ord_cleaned = order.dropna(subset=["order_purchase_timestamp"])
```

#order review dataset

```python
order_rev = pd.read_csv('olist_order_reviews_dataset.csv')
print(order_rev.head())
print(order_rev.info())
```

```python
for col in ["review_creation_date", "review_answer_timestamp"]:
    order_rev[col] = pd.to_datetime(order_rev[col], errors="coerce")
```

```python
order_rev["review_score"] = order_rev["review_score"].astype(int) #make sure review score is int
```

```python
print(order_rev.isna().sum())
```

```python
order_rev_clean = order_rev.drop(columns=["review_comment_title", "review_comment_message"])
```

```python
order_rev_clean = order_rev_clean.drop_duplicates(subset="review_id")
```

```python
order_rev_clean.head(5)
```

```python
order_rev_clean = order_rev_clean.sort_values(by='review_score', ascending=False)
order_rev_clean.head()
```
